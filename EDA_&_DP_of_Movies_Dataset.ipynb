* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## The best and the worst movies...

2. __Filter__ the Dataset and __find the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

__Define__ an appropriate __user-defined function__ to reuse code.

```python
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
#create a separate object with most important topics to answer our questions
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
df_best
```

```python
#Feature Engineer
#2 additional columns that express the profits & return
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
```

```python
#Value of 1 rev = bud
#value < 1 = rev smaller than bud
df_best
```

```python
#change column labels for presentation purposes
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes",
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
df_best.set_index("Title", inplace = True)
```

```python
df_best
```

```python
df_best.iloc[0,0]
```

```python
subset = df_best.iloc[:5, :2]
subset
```

```python
HTML(subset.to_html(escape=False))
```

```python
#Naive approach to highest rated movie
df_best.sort_values(by = "Average Rating", ascending = False)
```

```python
df_best.sort_values(by = "ROI", ascending = False)
```

```python
df_best.loc[df_best.Budget >= 5].sort_values(by = "ROI", ascending = False)
```

```python
df_best.Budget.fillna(0, inplace = True)
df_best.Votes.fillna(0, inplace = True)
```

```python
df_best.info()
```

```python
#Function to display best movies
def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):

    #Take df_best filter out for minimum budget & votes
    #Select path of interest revenue/profit, sort datafram by criteria
    #get the best (n) movies
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes),
                      ["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    #return in HTML format
    return HTML(df2.to_html(escape=False))
```

__Movies Top 5 - Highest Revenue__

```python
best_worst(n = 5, by = "Revenue")
```

## Most Successful Actors

Goal is to create a dataframe featuring the most successful actors in terms of revenue, voting average & popularity

```python
df.cast
```

```python
df
```

index is set to "id" column as each "id" is a unique identifier code for the cast of the film

```python
df.set_index("id", inplace = True)
```

```python
df.info()
```

```python
df.cast
```

```python
act = df.cast.str.split("|", expand = True)
act
```

```python
#remove inner index level 
act = act.stack().reset_index(level=1, drop=True).to_frame()
act
```

```python
act.columns = ["Actor"]
```

```python
act = act.merge(df[["title", "revenue_musd", "vote_average", "popularity"]],
                how = "left", left_index = True, right_index = True)
```

```python
act
```

```python
act.Actor.nunique()
```

```python
act.Actor.unique()
```

```python
#most active actors
act.Actor.value_counts().head(20)
```

```python
#visualize with bar plt
plt.figure(figsize = (12, 8))
act.Actor.value_counts().head(20).plot(kind='bar', fontsize = 15)
plt.title("Most Active Actors", fontsize = 20)
plt.ylabel("Number of Movies", fontsize = 15)
plt.show()
```

```python
#aggregate most successful actors
agg = act.groupby("Actor").agg(Total_Revenue = ("revenue_musd_x", "sum"), 
                               Mean_Revenue = ("revenue_musd_x", "mean"),
                               Mean_Rating = ("vote_average_x", "mean"), 
                               Mean_Pop = ("popularity_x", "mean"), 
                               Total_Movies = ("Actor", "count"))
```

```python
agg.nlargest(10, "Total_Movies")
```

```python
agg.nlargest(10, "Total_Revenue")
```

```python
plt.figure(figsize = (12, 8))
agg.Total_Revenue.nlargest(10).plot(kind='bar', fontsize = 15)
plt.title("Total Revenue",fontsize = 20)
plt.ylabel("Revenue (in MUSD)", fontsize = 15)
plt.show()
```
